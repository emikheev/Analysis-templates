```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
x=np.arange(0,100,1)
y=np.arange(0,100,1)
xx,yy=np.meshgrid(x,y)
zz=(np.sin(xx)+np.cos(yy))


mesh=plt.pcolormesh(xx,yy,zz,cmap='magma',vmin=zz.min(),vmax=zz.max())
plt.xlabel('x');plt.ylabel('z');
#mesh.norm=colors.LogNorm(0.1,zz.max())
cbar = plt.colorbar();cbar.set_label('z')
```
